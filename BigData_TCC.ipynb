# ***Filtrando Dados***

```python
###casos confirmados de morte na uti que ocuparam leitos por mais de 50 dias ---> retornou= 3336 rows
###são 3336 leitos ocupados por mais 50 dias,que não deram a solução da doença,pelo contrario a pessoa faleceu
df.loc[ (df['saidaConfirmadaObitos'] == 1.0) & (df['ocupacaoConfirmadoUti'] > 50.0) ]
```

```python
##quase igual a celula acima .Com a clausula 'OU' saida suspeita de hóbitos.E também reduzindo otempo de ocupação,ficando no patamar de 15 dias  --> AGORA FICA : 111757 rows
df.loc[ (df['saidaConfirmadaObitos'] == 1.0) | (df['saidaSuspeitaObitos'] == 1.0) & (df['ocupacaoConfirmadoUti'] > 15.0) ]
```

## **Vamos trabalhar somente com este ultimo dataframe que filtramos então criamos um novo**

```python
novo_df = df.loc[ (df['saidaConfirmadaObitos'] > 1) | (df['saidaSuspeitaObitos'] > 1) & (df['ocupacaoConfirmadoUti'] > 15.0) ] 
novo_df  ##só para imprimir novo dataframe
```

```python
###como o index do novo dataframe manteve o index dos dados originais,vamos colocar um novo index,para facilitar os trabalhos
novo_df = novo_df.reset_index()
novo_df  ###agora o index está ordenado -->ao lado do novo index está o da tabela original
```

```python
novo_df.loc[ novo_df['_id'].str.contains('qkQ') ] ###procurar determinado caratcter na coluna --> por exemplo uma parte da string do _id
```

```python
###AGORA VOU IMPORTAR RE regular expression uma funcionalidade do python e truncar algumas informações
import re
novo_df.loc[ novo_df['_id'].str.contains('81T2|HRI',flags= re.I , regex =True) ]  ###somente coloquei um trecho das strings de dois registros e ver se existe algum no dataframe
###flags= re.I , regex =True  re.I quer dizer que vai ignorar se éminuscula ou maiscula e o True habilita o regular expression
```

```python
###Outro tipo poderoso do REGex na clausula :retorna todos os municipioNotificacao que inicia com 'Santo' sequido por qualquer sequencia de caracteres
novo_df.loc[ novo_df['municipioNotificacao'].str.contains('^Santo[A-Z]*',flags= re.I , regex =True) ] 
###flags= re.I , regex =True  re.I quer dizer que vai ignorar se éminuscula ou maiscula e o True habilita o regular expression
```

```python
###COMO SAIU AINDA UNS CAMPOS NAN NA CLAUSULA ACIMA EU DELETEI ELES COM DROPNA()
novo_df = df.dropna(axis=0)
novo_df
```
